```python
import pandas as pa
import numpy as nu
import statsmodels.api as sm

# Import csv
df = pa.read_csv("Travel Pony Facebook.csv")
print(df.head(2))


# Question 1

# First, create the analytics 'cost per impression' by dividing Amount Spent / Impressions. Do this for each ad in your data, and save that as a column.

df['cost per impressions'] = (df['Amount Spent (USD)']/df['Impressions'])
print (df['cost per impressions'])

# When considering 'cost per impression', what day of the week works best? (What day is it cheapest to generate impressions) 
# What day works worst? (What day is most expensive)
# To answer this question, you're going to want to take the "Start Date" column and extract out the day of week. (Using arrow)

import arrow

SD = df['Start Date']
daysoftheweek = []

for row in range(len(df)):
    dt = arrow.get(df['Start Date'][row], "M/D/YY")
    daysoftheweek.append(dt.weekday())
    
df['Week Day'] = daysoftheweek
group = df.groupby('Week Day')
weekdaylist = group.aggregate(nu.mean)
lowestcpi = min(weekdaylist['Amount Spent (USD)']/weekdaylist['Impressions'])
cpi = weekdaylist['Amount Spent (USD)']/weekdaylist['Impressions']
print(min(cpi))
print(max(cpi))

# ANSWER #
# Tuesday is the the most cost effective day to generate impressions at $0.0021535 per impression on average. Contrastingly, Sunday is the most expensive option with an average CPI of $0.0058789.


# ------------------------------ #
# Question 2

# Next, compute the correlation between Amount Spent and the following variables:
# - Reach
# - Frequency 
# - Unique Clicks
# - Page Likes

# Which correlation is the strongest? What does that mean practically? (respond in a tweet or less)

reach = nu.corrcoef(df['Amount Spent (USD)'], df['Reach']) 
print(reach)
freq = nu.corrcoef(df['Amount Spent (USD)'], df['Frequency'])
print(freq)
unicli = nu.corrcoef(df['Amount Spent (USD)'],df['Unique Clicks'])
print(unicli)
pagelikes = nu.corrcoef(df['Amount Spent (USD)'], df['Page Likes'])
print(pagelikes)

# ANSWER #
# Unique Clicks has the highest correlation score at 0.88299318. This means that spending more money is highly likely to have a positive impact on the number of unique clicks an ad receives.


# ------------------------------ #
# Question 3

# Finally, perform a simple multiple regression analysis where Unique Clicks is the dependent variable and Reach and Frequency are the independent (predictor) variables.
# What variable most strongly predicts unique clicks? What does that mean practically? (respond in a tweet or less)

y=df['Unique Clicks']
x=df[['Reach', 'Frequency']]

model = sm.OLS(y,x).fit()
predictions = model.predict(x)
print (model.summary())

# ANSWER #
# Frequency has the largest predictive impact on Unique Clicks, with a coefficient of 3.6139. This means that for each additional delivery of the given ad, Unique Clicks will increase by an average of 3.6139 times.
```
